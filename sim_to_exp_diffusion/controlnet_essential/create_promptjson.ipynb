For mapping from Sim to Exp 

```python
"""This creates a json file that contains information about source and target files and blank prompt for all training image pairs
input is just the two folders containing the input and output images, change accordingly"""
```

```python
import os
import json

from cldm.config import BASE_FOLDER, SPECIFIC_FOLDER_SIM, SPECIFIC_FOLDER_EXP , SPECIFIC_FOLDER_SEED

# Define dataset paths

base_folder=BASE_FOLDER

folder_i=SPECIFIC_FOLDER_SIM
folder_o=SPECIFIC_FOLDER_EXP

source_dir = os.path.join(base_folder, folder_i)
target_dir = os.path.join(base_folder, folder_o)
json_path = os.path.join(base_folder, "prompt_simtoexp.json")

# Get sorted lists of filenames
source_files = sorted(os.listdir(source_dir))
target_files = sorted(os.listdir(target_dir))

# Ensure the number of files match
assert len(source_files) == len(target_files), "Mismatch between source and target images!"

# Generate the prompt.json file (line-by-line format)
with open(json_path, "w") as f:
    for src, tgt in zip(source_files, target_files):
        entry = {
            "source": f"{src}",
            "target": f"{tgt}",
            "prompt": ""  # Empty text prompt
        }
        
        f.write(json.dumps(entry) + "\n")  # Write each entry as a new line

print(f"Generated {json_path} with {len(source_files)} entries.")
```
